### Using sympy for summations

```python
i, n = symbols("i n")

# iterate each element i from 1 to n
# then multiply each element by 2 and sum up
summation = Sum(2 * i, (i, 1, n))

# specify n as 5,
# iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
up_to_5.doit()  # 30
```

### Eulers Number

A property of Eulers number is its exponential function is a derivative to itself,
which is convenient for exponential and logarithmic functions. 

Example: Calculating compound interest in Python

```python
p = 100  # tarting investment P
r = .20  # interest rate r
t = 2.0  # time span
n = 12  # periods n  (compounded monthly)

a = p * (1 + r / n) ** (n * t)
a
```

## Limits

```python
x = symbols("x")
f = 1/x

plot(f)
```

Notice that as x forever increases, f(x) gets
closer to 0. Fascinatingly, f(x) never actually reaches 0. It just forever keeps getting
closer.

### Using SymPy to calculate limits

```python
# calulating the limit of f(x)=1/x as x goes to infinity
x = symbols("x")
f = 1/x

result = limit(f, x, oo)
print(result) 
```

```python
n = symbols("n")
f = (1 + (1/n))**n

result = limit(f, n, oo)

print(result) # eulers number
print(result.evalf()) # eulers number as a float
```
